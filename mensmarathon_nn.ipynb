```python
import pandas as pd
import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("data/athlete_events.csv")
# drop cols which have NA for suggusted features
df = df.dropna(subset=['Age', 'Height', 'Weight'])
# filter for just the men's marathon
df_mens_marathon =  df[df["Event"] == "Athletics Men's Marathon"]
# replace the NaN values from medal with "No Medal"
df_mens_marathon["Medal"] = df_mens_marathon["Medal"].fillna("No Medal")
# show head
df_mens_marathon.head()
```

```python
df_mens_marathon.info()
```

```python
df_mens_marathon["Medal"].value_counts()
```

```python
# split into features and output
y = df_mens_marathon["Medal"]
X = df_mens_marathon[["Age", "Height", "Weight", "Year"]]

# test train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)
```

```python
y_pred = mlp.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy*100:.2f}%")
```
